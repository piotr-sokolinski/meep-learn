```python
import meep as mp

cell = mp.Vector3(16,8,0)
```

```python
geometry = [
    mp.Block(
        mp.Vector3(mp.inf, 1, mp.inf),
        center=mp.Vector3(),
        material=mp.Medium(epsilon=12),
    )
]
```

```python
sources = [
    mp.Source(
        mp.ContinuousSource(frequency=0.15), component=mp.Ez, center=mp.Vector3(-7, 0)
    )
]
```

```python
pml_layers = [mp.PML(1.0)]
```

```python
resolution = 10
```

```python
sim = mp.Simulation(
    cell_size=cell,
    boundary_layers=pml_layers,
    geometry=geometry,
    sources=sources,
    resolution=resolution,
)
```

```python
from matplotlib import pyplot as plt

%matplotlib inline
plt.figure(dpi=100)
sim.plot2D()
plt.show()
```

```python
sim.run(until=200)

```

```python
plt.figure(dpi=100)
sim.plot2D(fields=mp.Ez)
plt.show()
```
